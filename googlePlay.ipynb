```python
from pandas.core.tools.numeric import to_numeric
from pandas.io.formats.style_render import Subset
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

play = pd.read_csv('apps.csv')
play.isna().values.sum()
play.shape
play.sample(5)
play.drop(['Last_Updated', 'Android_Ver'], axis=1, inplace=True)
df_apps_clean = play.dropna()
df_apps_clean.head()
df_apps_clean.shape
df_apps_clean.duplicated()
apps_not_duplicate = df_apps_clean.drop_duplicates()
df_apps_clean= df_apps_clean.drop_duplicates(subset=['App', 'Type', 'Price'])
df_apps_clean[df_apps_clean.App=='Instagram']
df_apps_clean.shape
df_apps_clean.App[df_apps_clean.Rating.idxmax()]
df_apps_clean.App[df_apps_clean.Size_MBs.idxmax()]
re = df_apps_clean.sort_values('Reviews', ascending=False).head(50)
re['Type']!='Free'
rating = df_apps_clean.Content_Rating.value_counts()
rating
fig = px.pie(labels=rating.index, values=rating.values, names=rating.index)
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
df_apps_clean.Installs.describe()
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(',',"")
df_apps_clean.Installs = to_numeric(df_apps_clean.Installs)
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$',"")
df_apps_clean.Price = to_numeric(df_apps_clean.Price)
df_apps_clean.sort_values('Price', ascending=False).head(50)
apps_less = df_apps_clean[df_apps_clean['Price']<250]
apps_less.sort_values('Price', ascending=False).head(10)
df_apps_clean['Revenue_Estimate'] = df_apps_clean.Installs.multiply(df_apps_clean.Price)
a = df_apps_clean.sort_values('Revenue_Estimate', ascending=False).head(10)
a['Category']=='GAME'
df_apps_clean.Category.nunique()
top_10 = df_apps_clean.Category.value_counts(10)
bar = px.bar(x=top_10.index, y=top_10.values)
bar.show()
category_installs = df_apps_clean.groupby('Category').agg({'Installs':pd.Series.sum})
category_installs.sort_values('Installs', ascending=True, inplace=True)
h_bar = px.bar(x=category_installs.Installs, y=category_installs.index, orientation='h')
h_bar.show()
df = df_apps_clean.groupby('Category').agg({'App':pd.Series.count})
categ_merged = pd.merge(df, category_installs, on='Category', how='inner')
categ_merged
scatter = px.scatter(categ_merged, x='App', y='Installs', title='Category Concentration', size='App', color='Installs', hover_name=categ_merged.index)
scatter.show()
df_apps_clean.Genres.value_counts().sort_values(ascending=True)[:5]
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
genre = stack.value_counts()
gbar = px.bar(x=genre.index, y=genre.values, title='Top Genres', color=genre.values, color_continuous_scale='Agsunset', hover_name=genre.index)
gbar.update_layout(xaxis_title = 'Genre', yaxis_title='Number of Apps', coloraxis_showscale=False)
gbar.show()
df_apps_clean.Type.value_counts()
df_free_vs_paid = df_apps_clean.groupby(['Type', 'Category'], as_index=False).agg({'App':pd.Series.count})
paid_bar = px.bar(df_free_vs_paid, x='Category', y='App', title='Free vs Paid by Category', color='Type', barmode='group')
paid_bar.update_layout(xaxis={'categoryorder':'total descending'}, yaxis=dict(type='log'))
paid_bar.show()
ibar = px.box(df_apps_clean, x='Type', y='Installs', color='Type', notched=True, title='How many downloads do paid apps give?', points='all')
ibar.update_layout(yaxis=dict(type='log'))
ibar.show()
df_paid_apps = df_apps_clean[df_apps_clean['Type']=='Paid']
pbox = px.box(df_paid_apps, x='Category', y='Revenue_Estimate', title='How much can paid apps earn?')
pbox.update_layout(xaxis_title='Category', yaxis_title='Paid App Ballmark Revenue', xaxis={'categoryorder':'min ascending'}, yaxis=dict(type='log'))
pbox.show()
pri_cate = px.box(df_paid_apps, x='Category', y='Price', title='Price per Category')
pri_cate.update_layout(xaxis_title='Category', yaxis_title='Paid App Price', xaxis={'categoryorder':'max descending'}, yaxis=dict(type='log'))
pri_cate.show()
```
